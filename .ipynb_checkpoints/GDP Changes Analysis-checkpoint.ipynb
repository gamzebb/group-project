### See the GDP (Gross Domestic Product) Changes by Year

```python
# Import libraries.
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#check where the files are
#print("Current folder:", os.getcwd())

#for file in os.listdir():
#    print("File:", file)
```

```python
# Read the GDP rate file and create the data frame

df_monthly_gdp = pd.read_csv('cleaned_economic_data/gdp_df.csv')

df_monthly_gdp.head() 
```

```python
# Get min and max dates
min_date = df_monthly_gdp['Date'].min()
max_date = df_monthly_gdp['Date'].max()

print("Minimum date:", min_date)
print("Maximum date:", max_date)
```

```python
# Plot it for visualisation
plt.figure(figsize=(14, 5))
plt.plot(df_monthly_gdp['Date'], df_monthly_gdp['GDP'], color='blue', linewidth=1.5)
plt.title('UK Monthly GDP Time Series 1955-2024')
plt.xlabel('Date')
plt.ylabel('GDP (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Read the quarter on quarter, seasonally adjusted GDP rate file and create the data frame

df_quarterly_gdp = pd.read_csv('cleaned_economic_data/quarterly_adjusted_gdp_percent.csv')

df_quarterly_gdp.head() 
```

```python
# Add a new date column and strip out space
df_quarterly_gdp['quarter_clean'] = df_quarterly_gdp['quarter_end_date'].str.replace(' ', '', regex=False)

df_quarterly_gdp.head() 
```

```python
# Reformat quarterly clean to enable plotting

# Convert to PeriodIndex and then to timestamp
df_quarterly_gdp['quarter_end_date'] = pd.PeriodIndex(
    df_quarterly_gdp['quarter_clean'], freq='Q'
).to_timestamp(how='end')

df_quarterly_gdp.head() 
```

```python
# Get min and max dates
min_date = df_quarterly_gdp['quarter_end_date'].min()
max_date = df_quarterly_gdp['quarter_end_date'].max()

print("Minimum date:", min_date)
print("Maximum date:", max_date)
```

```python
# Plot it for visualisation
plt.figure(figsize=(14, 5))
plt.plot(df_quarterly_gdp['quarter_end_date'], df_quarterly_gdp['qoq _cvm _sa _percent'], color='blue', linewidth=1.5)
plt.title('UK quarterly GDP Time Series: 1955 - 2014')
plt.xlabel('Quarter')
plt.ylabel('GDP: chain volume measured, seasonaly adjusted (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Create the figure and axes
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8), sharex=False)

# First plot: Monthly GDP
ax1.plot(df_monthly_gdp['Date'], df_monthly_gdp['GDP'], color='blue')
ax1.set_title('Monthly GDP')
ax1.set_ylabel('Monthly GDP %')
ax1.set_xlabel('Date')
ax1.grid(True)

# Second plot: Quarterly GDP
ax2.plot(df_quarterly_gdp['quarter_end_date'], df_quarterly_gdp['qoq _cvm _sa _percent'], color='green')
ax2.set_title('Quarterly GDP (CVM SA %)')
ax2.set_ylabel('Quarterly GDP (CVM, SA %)')
ax2.set_xlabel('Date')
ax2.grid(True)

# Clean up layout
plt.tight_layout()
plt.show()
```

```python
# Read the speeches file and create the data frame

cleaned_speeches_uk = pd.read_excel('CleanedDataUK.xlsx')

cleaned_speeches_uk.head() 
```

```python
# create a dataframe with just the dates

date_speech_uk = cleaned_speeches_uk[['Date']]
date_speech_uk.columns = ['speech_date']

date_speech_uk.head() 
```

```python
# Create a copy
date_speech_uk = date_speech_uk.copy()

# Extract year and month
date_speech_uk['year_month'] = date_speech_uk['speech_date'].dt.to_period('M')

# Group by year and month, and count the number of speeches
monthly_speech_counts = date_speech_uk.groupby('year_month').size().reset_index(name='speech_count')

# Display the new DataFrame
monthly_speech_counts.head()
```

```python
#Check for data types
monthly_speech_counts.dtypes
```

```python
# Convert to datetime
monthly_speech_counts['year_month'] = monthly_speech_counts['year_month'].dt.to_timestamp()
```

```python
#Check for data types
monthly_speech_counts.dtypes
```

```python
#plot the speech dates on a timeline

# Start the plot
plt.figure(figsize=(14, 4))

# Plot
plt.figure(figsize=(14, 6))
plt.plot(monthly_speech_counts['year_month'], monthly_speech_counts['speech_count'], marker='o', linestyle='-')
plt.title('Monthly Count of Bank of England Speeches')
plt.xlabel('Date')
plt.ylabel('Number of Speeches')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter both Monthly and Quarterly GDP by 1997-2022

# Convert 'Date' column to datetime type
df_monthly_gdp['Date'] = pd.to_datetime(df_monthly_gdp['Date'])

# Monthly GDP
df_monthly_gdp_filtered = df_monthly_gdp[
    (df_monthly_gdp['Date'] >= '1997-01-01') &
    (df_monthly_gdp['Date'] <= '2022-12-31')
]


# Quarterly GDP (CVM, SA %)
df_quarterly_gdp_filtered = df_quarterly_gdp[
    (df_quarterly_gdp['quarter_end_date'] >= '1997Q1') &
    (df_quarterly_gdp['quarter_end_date'] <= '2022Q4')
]
# Convert 'quarter_end_date' to datetime (last day of quarter)
df_quarterly_gdp_filtered['quarter_date'] = pd.PeriodIndex(df_quarterly_gdp_filtered['quarter_end_date'], freq='Q').to_timestamp(how='end')


# Reset index
df_monthly_gdp_filtered = df_monthly_gdp_filtered.reset_index(drop=True)
df_quarterly_gdp_filtered = df_quarterly_gdp_filtered.reset_index(drop=True)


print("\n==== Monthly GDP (Filtered: 19972022) ====")
print(df_monthly_gdp_filtered)

print("==== Quarterly GDP (Filtered: 19972022) ====")
print(df_quarterly_gdp_filtered)
```

```python
#Min and Max date check

# Monthly GDP
min_date = df_monthly_gdp_filtered['Date'].min()
max_date = df_monthly_gdp_filtered['Date'].max()

print("Minimum Monthly GDP date:", min_date)
print("Maximum Monthly GDP date:", max_date)

# Quarterly GDP
min_date = df_quarterly_gdp_filtered['quarter_end_date'].min()
max_date = df_quarterly_gdp_filtered['quarter_end_date'].max()

print("Minimum Quarterly GDP date:", min_date)
print("Maximum Quarterly GDP date:", max_date)


# Monthly speech date 
min_date = monthly_speech_counts['year_month'].min()
max_date = monthly_speech_counts['year_month'].max()

print("Minimum Monthly speech date:", min_date)
print("Maximum Monthly speech date:", max_date)
```

```python
# Convert Period to datetime (timestamp at start of month)
monthly_speech_counts['year_month'] = monthly_speech_counts['year_month'].astype(str)
monthly_speech_counts['year_month'] = pd.to_datetime(monthly_speech_counts['year_month'])
```

```python
# Plot them all

# Create the figure and axes
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(14, 8), sharex=False)

# First plot: Monthly GDP
ax1.plot(df_monthly_gdp_filtered['Date'], df_monthly_gdp_filtered['GDP'], color='blue')

ax1.set_title('Monthly GDP')
ax1.set_ylabel('Monthly GDP %')
ax1.set_xlabel('Date')
ax1.grid(True)

# Second plot: Quarterly GDP
ax2.plot(df_quarterly_gdp_filtered['quarter_end_date'], df_quarterly_gdp_filtered['qoq _cvm _sa _percent'], color='green')
ax2.set_title('Quarterly GDP (CVM SA %)')
ax2.set_ylabel('Quarterly GDP (CVM, SA %)')
ax2.set_xlabel('Date')
ax2.grid(True)

# Third plot: Speeches a month (count)
ax3.plot(monthly_speech_counts['year_month'], monthly_speech_counts['speech_count'], color='red')
ax3.set_title('Monthly speech counts')
ax3.set_ylabel('Number of speeches')
ax3.set_xlabel('Date')
ax3.grid(True)

# Clean up layout
plt.tight_layout()
plt.show()
```
