**See the Bank Rate Changes by Year**

```python
# Import relevant libraries.
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Read the bank rate file and create the data frame

df_bank_rates = pd.read_excel('C:\\Users\\elif.ozkol\\Desktop\\LSE Project\\Bank Rate Changes.xlsx')

df_bank_rates 
```

```python
# Convert the Date column to datetime format
df_bank_rates ["Date"] = pd.to_datetime(df_bank_rates ["Date"])

# Create a new column 'Year' from the 'Date' column
df_bank_rates ["Year"] = df_bank_rates ["Date"].dt.year

df_bank_rates 
```

```python
# Group by 'Year' and calculate average 'BankRate'
grouped_bank_rates  = df_bank_rates .groupby("Year")["BankRate"].mean().reset_index()

grouped_bank_rates
```

```python
# See the change by year

# Plot the data
plt.figure(figsize=(14, 6))
plt.plot(grouped_bank_rates["Year"], grouped_bank_rates["BankRate"], marker='o')
plt.title("Average Bank Rate per Year")
plt.xlabel("Year")
plt.ylabel("Average Bank Rate (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Observations and possible reasons of the changes:**
- A sharp drop around 2009 during the global financial crisis.
- Very low rates sustained through the 2010s.
- A steep rise starting in 2022, likely in response to inflation.
