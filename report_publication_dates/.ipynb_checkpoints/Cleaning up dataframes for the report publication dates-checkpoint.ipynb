```python
import pandas as pd

# Load the uploaded Excel files
url1=  'https://github.com/gamzebb/group-project/raw/refs/heads/main/report_publication_dates/MPR_dates.xlsx'
monetary_policy_df = pd.read_excel(url1)

url2= 'https://github.com/gamzebb/group-project/raw/refs/heads/main/report_publication_dates/FSR_dates.xlsx'
financial_stability_df = pd.read_excel(url2)

monetary_policy_df.head(), financial_stability_df.head()
```

```python
# Process Monetary Policy Report dates
monetary_policy_dates = pd.melt(monetary_policy_df, id_vars=['Year'], value_vars=['Q1', 'Q2', 'Q3', 'Q4'])
monetary_policy_dates = pd.to_datetime(monetary_policy_dates['value'], errors='coerce')
monetary_policy_dates = monetary_policy_dates.dropna()
monetary_policy_dates = monetary_policy_dates[monetary_policy_dates >= pd.to_datetime('1998-09-01')]

# Process Financial Stability Report dates
# Flatten the entire DataFrame to one list, parse dates, and clean
flat_fsr_dates = pd.to_datetime(financial_stability_df.stack(), errors='coerce')
flat_fsr_dates = flat_fsr_dates.dropna()
flat_fsr_dates = flat_fsr_dates[flat_fsr_dates >= pd.to_datetime('1998-09-01')]

# Create cleaned DataFrames
mpr_dates_cleaned = pd.DataFrame({'date': monetary_policy_dates.sort_values().reset_index(drop=True)})
fsr_dates_cleaned = pd.DataFrame({'date': flat_fsr_dates.sort_values().reset_index(drop=True)})

# mpr_dates_cleaned.to_csv("mpr_dates_cleaned.csv", index=False)
# fsr_dates_cleaned.to_csv("fsr_dates_cleaned.csv", index=False)

mpr_dates_cleaned.head(), fsr_dates_cleaned.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Copy from cleaned data
mpr_df = mpr_dates_cleaned.copy()
fsr_df = fsr_dates_cleaned.copy()

# Key economic events
events = {
    'Early 1990s Recession (UK)': {'Start': '1990-07-01', 'End': '1992-09-30'},
    'Black Wednesday (ERM Crisis)': {'Start': '1992-09-16', 'End': '1992-09-16'},
    'BoE Operational Independence': {'Start': '1997-05-06', 'End': '1997-05-06'},
    'Dot-com Bubble Burst': {'Start': '2000-03-01', 'End': '2002-10-31'},
    '9/11 Attacks': {'Start': '2001-09-11', 'End': '2001-09-11'},
    'Global Financial Crisis (GFC)': {'Start': '2007-08-01', 'End': '2009-03-31'},
    'Lehman Brothers Collapse': {'Start': '2008-09-15', 'End': '2008-09-15'},
    'Eurozone Debt Crisis': {'Start': '2010-05-01', 'End': '2012-08-31'},
    'Brexit Referendum': {'Start': '2016-06-23', 'End': '2016-06-23'},
    'Brexit Negotiation Period': {'Start': '2016-07-01', 'End': '2019-12-31'},
    'COVID-19 Crash & Lockdown': {'Start': '2020-03-01', 'End': '2020-05-31'},
    'Post-COVID Reopening Boom': {'Start': '2021-03-01', 'End': '2021-12-31'},
    'UK Inflation Shock / Energy Crisis': {'Start': '2022-01-01', 'End': '2022-12-31'},
    'UK Mini-Budget Gilt Crisis': {'Start': '2022-09-23', 'End': '2022-10-14'}
}

# Ensure datetime type
mpr_df['date'] = pd.to_datetime(mpr_df['date'], errors='coerce')
mpr_df = mpr_df.dropna(subset=['date'])
mpr_df['year'] = mpr_df['date'].dt.to_period('Y').dt.to_timestamp()
mpr_annual = mpr_df.groupby('year').size().reset_index(name='MPR')
mpr_annual.rename(columns={'year': 'Year'}, inplace=True)

fsr_df['date'] = pd.to_datetime(fsr_df['date'], errors='coerce')
fsr_df = fsr_df.dropna(subset=['date'])
fsr_df['year'] = fsr_df['date'].dt.to_period('Y').dt.to_timestamp()
fsr_annual = fsr_df.groupby('year').size().reset_index(name='FSR')
fsr_annual.rename(columns={'year': 'Year'}, inplace=True)

# Merge
publication_df = pd.merge(mpr_annual, fsr_annual, on='Year', how='outer').set_index('Year').sort_index().fillna(0)

# Plot
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(publication_df.index, publication_df['MPR'], label='Monetary Policy Report (MPR)', marker='o', color='blue')
ax.plot(publication_df.index, publication_df['FSR'], label='Financial Stability Report (FSR)', marker='s', color='green')

# Economic event shading
for event, dates in events.items():
    start = pd.to_datetime(dates['Start'])
    end = pd.to_datetime(dates['End'])
    if end >= publication_df.index.min():
        ax.axvspan(max(start, publication_df.index.min()), end, alpha=0.3, color='orange', label=event)

# Formatting
ax.set_title("Annual Frequency of MPR and FSR Publications with Economic Events")
ax.set_ylabel("Number of Reports Published")
ax.set_xlabel("Year")
ax.set_xlim(publication_df.index.min(), publication_df.index.max())
ax.xaxis.set_major_locator(mdates.YearLocator(2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)

# Deduplicate legend
handles, labels = ax.get_legend_handles_labels()
unique = dict(zip(labels, handles))
ax.legend(unique.values(), unique.keys(), loc='upper left', fontsize=8)

plt.grid(True, which='major', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Plot setup
fig, ax = plt.subplots(figsize=(14, 7))

# Lines
ax.plot(publication_df.index, publication_df['MPR'], label='Monetary Policy Report (MPR)', marker='o', color='blue')
ax.plot(publication_df.index, publication_df['FSR'], label='Financial Stability Report (FSR)', marker='s', color='green')

# Event shading + vertical text
for event, dates in events.items():
    start = pd.to_datetime(dates['Start'])
    end = pd.to_datetime(dates['End'])

    if end >= publication_df.index.min():
        shade_start = max(start, publication_df.index.min())
        ax.axvspan(shade_start, end, alpha=0.3, color='orange', zorder=1)

        # Label positioning
        label_x = shade_start + (end - shade_start) / 2
        ax.text(label_x, 0.08, event.split('(')[0].strip(), 
                rotation=90, fontsize=8, color='black', fontweight='bold',
                ha='left', va='bottom', transform=ax.get_xaxis_transform(),
                bbox=dict(facecolor='white', alpha=1, edgecolor='none', boxstyle='round,pad=0.2'))

# Formatting
ax.set_title("Annual Frequency of MPR and FSR Publications with Economic Events")
ax.set_ylabel("Number of Reports Published")
ax.set_xlabel("Year")
ax.set_xlim(publication_df.index.min(), publication_df.index.max())
ax.xaxis.set_major_locator(mdates.YearLocator(2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)

# Clean legend (only MPR/FSR)
handles, labels = ax.get_legend_handles_labels()
filtered = [(h, l) for h, l in zip(handles, labels) if 'Report' in l]
ax.legend(*zip(*filtered), loc='lower left', fontsize=8)

plt.grid(True, which='major', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
